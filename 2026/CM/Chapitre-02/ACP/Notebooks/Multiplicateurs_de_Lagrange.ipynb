## Solution théorique de l'optimisation sous contrainte

Soit une fonction $f: \mathbb{R}^p \to \mathbb{R}$ que l'on cherche à optimiser. Toutefois, on souhaite contraindre les valeurs de $x$ par une fonction $g: \mathbb{R}^p \to \mathbb{R}$ dont on fixe la valeur à $g(x) = 0$

$$
\begin{equation}
    \underset{\displaystyle x \in \mathbb{R}^p}{\operatorname{arg max}} \; f(x) \quad \text{sous contrainte} \quad g(x) = 0
\end{equation}
$$

Le théorème des multiplicateurs de Lagrange consiste à résoudre ce problème d'optimisation avec une contrainte d'égalité en introduisant le lagragien $\mathcal{L}$ et un multiplicateur $\lambda \in \mathbb{R}$:

$$
\begin{equation}
    \mathcal{L}(x, \lambda) = f(x) - \lambda g(x)
\end{equation}
$$

Résoudre le problème de maximisation de $f$ sous contrainte d'égalité $g(x) = 0$ consiste à trouver $(\lambda, x)$ tel que la dérivée du lagrangien $\mathcal{L}(x, \lambda)$ est nulle.

Pour annuler la dérivée du lagrangien, il faut à la fois chercher:

$$
\begin{equation}
    \begin{cases}
        \nabla_x \mathcal{L}(x, \lambda) = 0 & \quad \text{condition de stationnarité} \\
        \nabla_{\lambda} \mathcal{L}(x, \lambda) = 0 & \quad \text{condition d'admissibilité}
    \end{cases}
\end{equation}
$$

Considérons les fonctions $f: \mathbb{R}^2 \to \mathbb{R}$ et $g: \mathbb{R}^2 \to \mathbb{R}$ suivantes:

$$
\begin{equation}
    \begin{cases}
        f(x) &= -x_1^2 - 2 x_2^2 + 6 \\
        g(x) &= x_1^2 + x_2^2 - 1
    \end{cases}
\end{equation}
$$

La condition de stationnarité donne:

$$
\begin{equation}
    \begin{aligned}
        \nabla_x \mathcal{L}(x, \lambda) 
        &= \nabla_x f(x) - \nabla_x (\lambda g(x)) = 0 \\
        &= \begin{pmatrix}-2 x_1 \\ -4 x_2\end{pmatrix} - \lambda \begin{pmatrix}2 x_1 \\ 2 x_2\end{pmatrix} = 0
    \end{aligned}
\end{equation}
$$

La condition d'admissibilité donne:

$$
\begin{equation}
    \begin{aligned}
        \nabla_{\lambda} \mathcal{L}(x, \lambda) 
        &= \nabla_{\lambda} f(x) - \nabla_{\lambda} (\lambda g(x)) = 0 \\
        &= 0 - g(x) = 0 \\
        &= x_1^2 + x_2^2 - 1 = 0
    \end{aligned}
\end{equation}
$$

Techniquement, la condition d'admissibilité est la vérification que la contrainte imposée est bien satisfaite !

Cela nous amène à un système non-linéaire de trois équations à trois inconnues:

$$
\begin{equation}
    \begin{cases}
        -2 x_1 - 2 \lambda x_1 = 0 \\
        -4 x_2 - 2 \lambda x_2 = 0 \\
        x_1^2 + x_2^2 - 1 = 0
    \end{cases}
\end{equation}
$$

Qui est équivalent à:

$$
\begin{equation}
    \begin{cases}
        x_1 (\lambda + 1) = 0 \\
        x_2 (\lambda + 2) = 0 \\
        x_1^2 + x_2^2 - 1 = 0
    \end{cases}
\end{equation}
$$

En partant de la première équation:

$$
\begin{equation}
    x_1 (\lambda + 1) = 0 \rightarrow x_1 = 0 \; \text{ou} \; \lambda = -1
\end{equation}
$$

En injectant $\lambda = -1$ dans la seconde équation:

$$
\begin{equation}
    x_2 (\lambda + 2) = 0 \Rightarrow x_2 = 0
\end{equation}
$$

En injectant $x_2 = 0$ dans la dernière équation:

$$
\begin{equation}
    x_1^2 + x_2^2 - 1 = 0 \Rightarrow x_1^2 - 1= 0 \Rightarrow x_1 = \pm 1
\end{equation}
$$

Ce qui conduit à deux premières solutions pour le système:

$$
\begin{equation}
    \begin{cases}
        x_1 = 1, & x_2 = 0, & \lambda = -1\\
        x_1 = -1, & x_2 = 0, & \lambda = -1
    \end{cases}
\end{equation}
$$

Partons maintenant de la deuxième équation:

$$
\begin{equation}
    x_2 (\lambda + 2) = 0 \rightarrow x_2 = 0 \; \text{ou} \; \lambda = -2
\end{equation}
$$

En injectant $\lambda = -2$ dans la première équation:

$$
\begin{equation}
    x_1 (\lambda + 1) = 0 \Rightarrow x_1 = 0
\end{equation}
$$

En injectant $x_1 = 0$ dans la troisième équation:

$$
\begin{equation}
    x_1^2 + x_2^2 - 1 = 0 \Rightarrow x_2^2 - 1= 0 \Rightarrow x_2 = \pm 1
\end{equation}
$$

Cela conduit à deux nouvelles solutions:

$$
\begin{equation}
    \begin{cases}
        x_1 = 0, & x_2 = 1, & \lambda = -2 \\
        x_1 = 0, & x_2 = -1, & \lambda = -2
    \end{cases}
\end{equation}
$$

Afin de sélectionner les bonnes solutions, nous allons maintenant évaluer $f$ sur ces différentes solutions:

$$
\begin{equation}
    \begin{cases}
        f(x_1 = 1, x_2 = 0) = 5  \\
        f(x_1 = -1, x_2 = 0) = 5 \\
        f(x_1 = 0, x_2 = 1) = 4  \\
        f(x_1 = 0, x_2 = -1) = 4
    \end{cases}
\end{equation}
$$

On en déduit donc que $(x_1 = 1, x_2 = 0, \lambda = -1)$ et $(x_1 = -1, x_2 = 0, \lambda = -1)$ sont les deux solutions à notre problème de maximisation sous contrainte.

## Représentation graphique de l'optimisation sous contrainte

```python
def f(x1, x2):
    return -x1**2 - 2 * x2**2 + 6
```

```python
def g(x1, x2):
    return x1**2 + x2**2 - 1
    # return np.abs(x1) + np.abs(x2) - 1
    # return np.maximum(np.abs(x1), np.abs(x2)) - 1
```

```python
x1, x2 = np.meshgrid(np.linspace(-3., 3., 200), np.linspace(-2., 2., 200))
```

```python
z = f(x1, x2)
c = g(x1, x2)
```

```python
levels=[-2, -1, 0, 1, 2, 3, 4, 4.5, 5, 5.5, 6]
cont_f = plt.contour(x1, x2, z, levels=levels, linestyles="-")
cont_g = plt.contour(x1, x2, c, levels=[0], linestyles=":")
handles_f, _ = cont_f.legend_elements()
handles_g, _ = cont_g.legend_elements()
plt.clabel(cont_f);
plt.scatter([1, -1], [0, 0] , color="red", marker="+", zorder=10, label=r"Maximum de $f$ sous contrainte");
plt.scatter([0, -0], [-1, 1] , color="blue", marker="+", zorder=10, label=r"Minimum de $f$ sous contrainte");
qf = plt.quiver(1, 0, -1, 0, color="green", angles='xy', scale_units='xy', scale=2.0, width=0.005, label=r"$\nabla_x f(x)$")
qg = plt.quiver(1, 0, 1, 0, color="black", angles='xy', scale_units='xy', scale=2.0, width=0.005, label=r"$\nabla_x g(x)$")
qf = plt.quiver(-1, 0, 1, 0, color="green", angles='xy', scale_units='xy', scale=2.0, width=0.005)
qg = plt.quiver(-1, 0, -1, 0, color="black", angles='xy', scale_units='xy', scale=2.0, width=0.005)
handles, labels = plt.gca().get_legend_handles_labels()
handles.extend([handles_f[0], handles_g[0]])
labels.extend([r"iso-contours de $f(x)$", r"$g(x) = 0$"])
plt.legend(handles, labels, loc="center left", bbox_to_anchor=(1.05, 0.5));
plt.axis("equal");
```

Les extremums sont les points de tangence des iso-contour de $f$ avec la contrainte $g$. On voit que la contrainte circulaire tangeante 2 iso-contour $f(x_1, x_2) = 5$ et $f(x_1, x_2) = 4$. Nous avons donc 2 maximums et 2 minimums de $f$ sous contrainte.

On niveau des maximums, les vecteurs représentent les gradients $\nabla_x f(x)$ et $\nabla_x g(x)$. On voit bien qu'en ces points de tangence on a la condition de Lagrange avec $\lambda = -1$:

$$
\begin{equation}
    \nabla_x f(x) = \lambda \nabla_x g(x)
\end{equation}
$$
