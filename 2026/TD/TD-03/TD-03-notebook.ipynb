## Execice 3

```python
def phi(X):
    return np.hstack((np.ones_like(X), X))
```

```python
X = np.array([175., 181., 183., 189.]).reshape(-1, 1)
Y = np.array([95., 103., 101., 109.]).reshape(-1, 1)
```

```python
X.mean(), X.std(), Y.mean(), Y.std()
```

```python
Xs = (X - X.mean()) / X.std()
Xs
```

```python
Ys = (Y - Y.mean()) / Y.std()
Ys
```

```python
phi(Xs)
```

```python
beta_hat = np.linalg.inv(phi(Xs).T @ phi(Xs)) @ phi(Xs).T @ Ys
beta_hat
```
