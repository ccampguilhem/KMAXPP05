## Exercice 1

```python
X = np.array([190., 172., 186., 180.]).reshape(-1, 1)
Y = np.array([112., 85., 135., 92.]).reshape(-1, 1)
```

```python
def phi(X):
    ones = np.ones(X.shape[0]).reshape(-1, 1)
    return np.hstack((ones, X))
```

```python
X_tilde = phi(X)
```

```python
X_tilde.T @ X_tilde
```

```python
np.linalg.det(X_tilde.T @ X_tilde)
```

```python
beta_hat = np.linalg.inv(X_tilde.T @ X_tilde) @ X_tilde.T @ Y
beta_hat
```

```python
Xi = np.array([188., 200.]).reshape(-1, 1)
Xi_tilde = phi(Xi)
Yi_hat = Xi_tilde @ beta_hat
Yi_hat
```
