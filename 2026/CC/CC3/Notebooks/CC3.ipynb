```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
1 / 6
```

```python
dieA = np.array([1, 1, 1, 1, 1, 1]) / 6.
dieB = np.array([1.75, 1.5, 1.25, 0.75, 0.5, 0.25]) / 6.
```

```python
distA = sp.stats.multinomial(60, p=dieA)
distB = sp.stats.multinomial(60, p=dieB)
```

```python
distA.rvs(1)
```

```python
distB.rvs(1)
```

```python
distA = sp.stats.multinomial(5, p=dieA)
distB = sp.stats.multinomial(5, p=dieB)
```

```python
distB.rvs(1)
```
