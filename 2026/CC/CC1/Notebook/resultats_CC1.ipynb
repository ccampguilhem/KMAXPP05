## Traitement des résultats

```python
def traitement_par_sujet(resultats):
    sortie = []
    for question in resultats:
        sortie.append([int(x) for x in list(question)])
    return np.array(sortie).T
```

```python
sujet_B
```

```python
traitement_par_sujet(resultats_sujet_B)
```

```python
df = pd.concat([
    pd.DataFrame(traitement_par_sujet(resultats_sujet_A), columns=sujet_A).assign(sujet="A"),
    pd.DataFrame(traitement_par_sujet(resultats_sujet_B), columns=sujet_B).assign(sujet="B"),
    pd.DataFrame(traitement_par_sujet(resultats_sujet_C), columns=sujet_C).assign(sujet="C"),
]).reset_index(drop=True).assign(etudiant=lambda x: x.index)
```

```python
df
```

```python
df_melt = df.melt(id_vars=["etudiant", "sujet"], var_name="question", value_name="points")
```

## Notes des étudiants

```python
def note(data, top):
    data = data.sort_values(["points"], ascending=False).head(top)
    return data["points"].sum()
df_notes = df_melt.groupby(["etudiant", "sujet"], as_index=False).apply(note, top=10).rename(columns={None: "note"})
```

## Questions les plus dures

```python
def taux_reussite_question(data):
    n = data.shape[0]
    mask = data["points"] >= 2
    return data[mask].shape[0] / n * 100
(df_melt
    .groupby("question", as_index=False)
    .apply(taux_reussite_question)
    .rename(columns={None: "taux réussite"})
    .sort_values("taux réussite", ascending=True)
)
```

```python
def points_moyens_question(data):
    n = data.shape[0]
    return data["points"].shape[0] / n * 100
(df_melt
    .groupby("question", as_index=False)
    .agg({"points": "mean"})
    .sort_values("points", ascending=True)
)
```
