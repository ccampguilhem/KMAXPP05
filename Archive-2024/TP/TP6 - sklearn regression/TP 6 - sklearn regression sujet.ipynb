## Validation croisée 

La validation croisée permet d'éviter l'overfitting. 

```python
from sklearn.model_selection import cross_val_score
LR = LinearRegression()
print(np.mean(-1 * cross_val_score(LR, X, y, cv = 5, scoring = 'neg_mean_squared_error')))
```
